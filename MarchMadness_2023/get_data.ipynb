```python
#load data from march_madness_pre.db into a dataframe
import pandas as pd
import sqlite3
conn = sqlite3.connect('march_madness.db')
c = conn.cursor()
df = pd.read_sql_query("SELECT * FROM march_madness_team_pts", conn)
df
```

```python
pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', 20)
df_query = pd.read_sql_query('''SELECT 
                        Josh_pts, Brant_pts, Justin_pts, Jeremy_pts, Joe_pts, Devan_pts, Nick_pts,
                        r16_w_game1, r16_w_game1_blowout,
                        r16_w_game2, r16_w_game2_blowout,
                        r16_s_game1, r16_s_game1_blowout,
                        r16_s_game2, r16_s_game2_blowout,
                        r16_e_game1, r16_e_game1_blowout,
                        r16_e_game2, r16_e_game2_blowout,
                        r16_mw_game1, r16_mw_game1_blowout,
                         r16_mw_game2, r16_mw_game2_blowout,
                        east_winner, east_winner_blowout,
                        south_winner, south_winner_blowout,
                        west_winner, west_winner_blowout,
                        midwest_winner, midwest_winner_blowout,
                        south_east_winner, south_east_winner_blowout,
                        west_midwest_winner, west_midwest_winner_blowout,
                        final_winner, final_winner_blowout
                        from march_madness
                             where final_winner not in ("Tennessee", "Florida Atlantic")
                             and Josh_win = 1
                             ''', conn)
#round values in df_query to 2 decimal places
df_query = df_query.round(2)
df_query
```
