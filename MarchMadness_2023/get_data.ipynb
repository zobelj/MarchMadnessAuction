```python
#load data from march_madness_pre.db into a dataframe
import pandas as pd
import sqlite3
conn = sqlite3.connect('march_madness_pre.db')
c = conn.cursor()
df = pd.read_sql_query("SELECT * FROM march_madness_pre", conn)
df
```

```python
df_query = pd.read_sql_query('''SELECT 
                    avg(Devan_win) DevWP, avg(Jeremy_win) JerWP, avg(Joe_win) JoeWP, avg(Brant_win) BraWP, avg(Justin_win) JusWP, avg(Josh_win) JoshWP, avg(Nick_win) NickWP,
                    avg(Devan_pts) DevPts, avg(Jeremy_pts) JerPts, avg(Joe_pts) JoePts, avg(Brant_pts) BraPts, avg(Justin_pts) JusPts, avg(Josh_pts) JoshPts, avg(Nick_pts) NickPts
                        FROM march_madness_pre
                             where east_winner = "Tennessee"''', conn)
#round values in df_query to 2 decimal places
df_query = df_query.round(2)
df_query
```

```python
joe_best = pd.read_sql_query('''SELECT 
                        r16_w_game1, r16_w_game2,
                        r16_s_game1, r16_s_game2,
                        r16_e_game1, r16_e_game2,
                        r16_mw_game1, r16_mw_game2,
                        east_winner, south_winner, west_winner,  midwest_winner, 
                        south_east_winner, west_midwest_winner, final_winner,
                        Devan_Pts
                        FROM march_madness_pre
                        where Devan_pts = (select max(Devan_pts) from march_madness_pre)''', conn)
#round values in df_query to 2 decimal places
joe_best
```
